```python
import requests
from bs4 import BeautifulSoup
import string
import pandas as pd
import time

#declare variables
df = pd.DataFrame()
combined_letters = []

#create list of every combination of two letter string. (aa,ab,...,zy,zz)
letters = list(string.ascii_lowercase)
for lettera in letters:
    for letterb in letters:
        combined_letters.append(lettera+letterb)

#for each combination send a post request to the server, scrape for the results, and store in a dataframe
for combination in combined_letters:
    #pause to avoid having requests denied
    time.sleep(.5)

    #store post request, data = the payload from the browser's post request
    r = requests.post(f'https://redrock.clarkcountynv.gov/ccdcincustody/inCustodySearch.aspx', data={"__EVENTTARGET": "","__EVENTARGUMENT": "", "__VIEWSTATE": "/wEPDwUIOTE5NTg4NjhkZBeUBD5/Gemp1x1+UwLqZlyt8HTpwd8HU590qPO6YExc", "__VIEWSTATEGENERATOR": "B6734120", "__EVENTVALIDATION": "/wEdAAVYVELKTCPb7RS30jYlWkXHJNOHGRTiA26VmVCqyHoxqsWzilCqUZiLm2Tkt0sWoCWjOglmm5fGSk5UkpFsfK4enirgrpbTwEsK9BgCJuFqyXku3iz80Sz3o8QsyPBGSSeiCxAUGo0m5YWY/zLJBlev", "TxtID": "", "TxtCaseNo": "", "txtName": {combination}, "SearchName": "  Submit"})
    print(combination, ": ", r.status_code)

    #parse resulting html and store data table
    soup = BeautifulSoup(r.text, 'html.parser')
    tbl = soup.find("table",{"id":"DataGrid1"})

    #store data table as dataframe, then concatenate it with the main dataframe
    loop_df = pd.read_html(str(tbl), header=0)[0]
    df = pd.concat([df,loop_df], ignore_index = True)

    #print(loop_df)
    #print(df)

    print(combination, ": ", "Done!")

#save dataframe as csv
df.to_csv('clark_county.csv')
```
